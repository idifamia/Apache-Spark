```
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, TimestampType, DoubleType
```

```
myschema = StructType( [
  StructField("InvoiceNo",StringType()),\
  StructField("StockCode",StringType()),\
  StructField("Descripcion",StringType()),\
  StructField("Cantidad",IntegerType()),\
  StructField("InvoiceDate",TimestampType()),\
  StructField("PrecioUnidad",DoubleType()),\
  StructField("CustomerID",DoubleType()),\
  StructField("Country",StringType())
  ]
)
```

```
df_1 = spark.read.format("csv")\
    .option("header","true")\
    .schema(myschema)\
    .load(path_retail_byday+"/2010-12-01.csv")
```

```
df_1.show()
```

```
df_1.printSchema()
```

```
df_1.count()
```

```
# path_retail_byday+"/2010-12-01.csv"
df = spark.read.format("csv")\
    .option("header","true")\
    .schema(myschema)\
    .load(path_retail_byday)
```

```
df.show()
```

```
df.count()
```

```
df.createOrReplaceTempView("transactions")
```

```
spark.sql("SELECT InvoiceNo, Descripcion, cantidad * preciounidad as Total_Cost FROM transactions").show(5)
```

```
%sql
select * from transactions
```
